class MyQueue:

    def __init__(self):
        self.in_stack = []
        self.out_stack = []

    def push(self, x: int) -> None:
        self.in_stack.append(x)

    def pop(self) -> int:
        self.peek()  # out_stack이 비어있으면 채움
        return self.out_stack.pop()

    def peek(self) -> int:
        if not self.out_stack:
            while self.in_stack:
                self.out_stack.append(self.in_stack.pop())
        return self.out_stack[-1]

    def empty(self) -> bool:
        return not self.in_stack and not self.out_stack

# 04
연습문제
```
구조
push(x): 큐1에 그냥 넣기

pop():

큐1에서 하나 빼서 큐2에 넣기를 반복 (마지막 남은 1개 제외)

마지막 하나는 스택의 pop 대상

그리고 큐1과 큐2를 swap

예시 코드

from collections import deque

class StackUsingQueues:
    def __init__(self):
        self.q1 = deque()
        self.q2 = deque()

    def push(self, x):
        self.q1.append(x)

    def pop(self):
        # q1에서 마지막 요소만 남기고 q2로 옮기기
        while len(self.q1) > 1:
            self.q2.append(self.q1.popleft())
        
        popped = self.q1.popleft()  # 마지막 요소 pop

        # q1 <-> q2 바꾸기
        self.q1, self.q2 = self.q2, self.q1

        return popped

    def top(self):
        # pop 대신 peek만 하고 싶을 때
        while len(self.q1) > 1:
            self.q2.append(self.q1.popleft())
        
        top_item = self.q1.popleft()
        self.q2.append(top_item)

        self.q1, self.q2 = self.q2, self.q1

        return top_item

    def isEmpty(self):
        return len(self.q1) == 0
```
# 05
연습문제
```
구조
enqueue(x) 할 땐 → stack_in에 push

dequeue() 할 땐 →

stack_out이 비었으면 stack_in에 있는 걸 모두 꺼내서 stack_out으로 옮기기

그리고 stack_out에서 pop

예시 코드
class QueueUsingStacks:
    def __init__(self):
        self.stack_in = []   # 데이터를 넣는 스택
        self.stack_out = []  # 데이터를 꺼낼 때 쓰는 스택

    def enqueue(self, x):
        self.stack_in.append(x)

    def dequeue(self):
        if not self.stack_out:
            # stack_out이 비었으면 stack_in의 데이터를 옮긴다
            while self.stack_in:
                self.stack_out.append(self.stack_in.pop())
        if not self.stack_out:
            raise IndexError("Queue is empty")
        return self.stack_out.pop()

    def front(self):
        if not self.stack_out:
            while self.stack_in:
                self.stack_out.append(self.stack_in.pop())
        if not self.stack_out:
            raise IndexError("Queue is empty")
        return self.stack_out[-1]

    def isEmpty(self):
        return not self.stack_in and not self.stack_out
```

```
class ListDeque:
    def __init__(self):
        self._queue = []

    def isEmpty(self) -> bool:
        return len(self._queue) == 0

    def addFront(self, x):
        self._queue.insert(0, x)   # 앞에 삽입

    def addRear(self, x):
        self._queue.append(x)      # 뒤에 삽입

    def deleteFront(self):
        if self.isEmpty():
            return None
        return self._queue.pop(0)  # 앞에서 삭제

    def deleteRear(self):
        if self.isEmpty():
            return None
        return self._queue.pop()   # 뒤에서 삭제

    def getFront(self):
        if self.isEmpty():
            return None
        return self._queue[0]      # 앞에 있는 값 반환

    def getRear(self):
        if self.isEmpty():
            return None
        return self._queue[-1]     # 뒤에 있는 값 반환

    def dequeueAll(self):
        self._queue.clear()

    def printQueue(self):
        print("Deque from front:", end=' ')
        for item in self._queue:
            print(item, end=' ')
        print()

addFront(x)	앞에 원소 삽입
addRear(x)	뒤에 원소 삽입 (기존 enqueue)
deleteFront()	앞에서 삭제 (기존 dequeue)
deleteRear()	뒤에서 삭제
getFront()	앞에 있는 원소 조회
getRear()	뒤에 있는 원소 조회
```

## 문제 3. Heap을 이용한 생일 상위 10명 출력

```python
import pandas as pd
from heap import Heap

df = pd.read_csv("DS_Birthdaydata - 시트1.csv")
df['생년월일'] = pd.to_datetime(df['생년월일'], errors='coerce')

heap_data = [
    (row['생년월일'].timestamp(), row['이름'])
    for _, row in df.iterrows()
    if pd.notna(row['생년월일'])
]

h = Heap()
for item in heap_data:
    h.insert(item)

for _ in range(min(10, h.size())):
    timestamp, name = h.deleteMax()
    birthday = pd.to_datetime(timestamp, unit='s').strftime('%Y-%m-%d')
    print(f"{name} - {birthday}")
```

## 문제 6. LeetCode 703. Kth Largest Element in Stream

```python
import heapq

class KthLargest:

    def __init__(self, k: int, nums: list[int]):
        self.k = k
        self.heap = nums
        heapq.heapify(self.heap)

        while len(self.heap) > k:
            heapq.heappop(self.heap)

    def add(self, val: int) -> int:
        heapq.heappush(self.heap, val)
        if len(self.heap) > self.k:
            heapq.heappop(self.heap)
        return self.heap[0]
```

```python
kth = KthLargest(3, [4, 5, 8, 2])
print(kth.add(3))  # 출력: 4
print(kth.add(5))  # 출력: 5
print(kth.add(10)) # 출력: 5
print(kth.add(9))  # 출력: 8
print(kth.add(4))  # 출력: 8
```
